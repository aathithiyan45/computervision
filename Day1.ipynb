```python
import cv2
from matplotlib import pyplot as plt

# Load the image
image = cv2.imread('Z:/test1.jfif') 

# Check if image was loaded
if image is None:
    print(" Error: Could not load image. Check the path and file name.")
else:
    # Convert to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Display the result
    plt.imshow(gray_image, cmap='gray')
    plt.axis('off')
    plt.show()
```

```python
import cv2

# Load the image
image = cv2.imread('Z:/test3.jpg') 

# Check if image was loaded
if image is None:
    print(" Error: Could not load image. Check the path and file name.")
else:
    # Apply Gaussian blur to the original image
    blurred = cv2.GaussianBlur(image, (5, 5), 0)

    # Create resizable windows
    cv2.namedWindow('Original Image', cv2.WINDOW_NORMAL)
    cv2.namedWindow('Blurred Image', cv2.WINDOW_NORMAL)

    # Show images in resizable windows
    cv2.imshow('Original Image', image)
    cv2.imshow('Blurred Image', blurred)

    # Wait for a key press and then close the windows
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
import cv2
from matplotlib import pyplot as plt

# Load the image
image = cv2.imread('Z:/test1.jfif') 

# Check if image was loaded
if image is None:
    print(" Error: Could not load image. Check the path and file name.")
else:
    # Convert BGR to RGB
    rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Display the RGB image
    plt.imshow(rgb_image)
    plt.axis('off')
    plt.title('RGB Image')
    plt.show()

    # Example: Print RGB value of pixel at (100, 100)
    pixel_value = rgb_image[100, 100]
    print(f" RGB value at (100, 100): {pixel_value}")
```

```python

import cv2
import numpy as np

# Load the image in grayscale
gray_image = cv2.imread('Z:/man-8554081_1280.jpg', cv2.IMREAD_GRAYSCALE)

# Check if image was loaded
if gray_image is None:
    print(" Error: Could not load image.")
else:
    # Example: Set pixels with intensity < 50 to white (255)
    gray_image[gray_image < 50] = 255

    # Show the modified image
    cv2.imshow('Modified Image', gray_image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
import cv2
from matplotlib import pyplot as plt

# Load the image
image = cv2.imread('Z:/test1.jfif') 

# Check if image was loaded
if image is None:
    print(" Error: Could not load image. Check the path and file name.")
else:
    # Print image shape
    height, width, channels = image.shape
    print(f"Height: {height}")
    print(f"Width: {width}")
    print(f"Channels: {channels}")

    # Convert to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Display the grayscale image
    plt.imshow(gray_image, cmap='gray')
    plt.axis('off')
    plt.show()
```
