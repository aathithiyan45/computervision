```python
import cv2
import numpy as np
# Open a video capture
cap = cv2.VideoCapture('test.mp4')  # 0 for webcam, or replace with video file path
no_frames = 0
while True:
    ret, frame = cap.read()
    no_frames =  no_frames +1 
    if not ret:
        break
    
    # Convert frame to grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    
    cv2.rectangle(frame,(110,50),(250,300),(0,120,255),3)
    #cv2.rectangle(frame,(110,50),(250,300),(120,234,255),cv2.FILLED)
    
    
    cv2.circle(frame,(400,100),50,(125,55,89),5)
    #cv2.circle(frame,(400,100),50,(255,123,35),cv2.FILLED)
    
    cv2.imshow("video_frames", frame)
    cv2.imshow("video_frames_gray", gray)
    
    # Exit on pressing 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
# Release resources
print(no_frames)
cap.release()
cv2.destroyAllWindows()
```

```python
import torch
import torch.nn as nn

# Sample data
x = torch.tensor([[1.0], [2.0], [3.0]])
y = torch.tensor([[2.0], [4.0], [6.0]])

# Define model
model = nn.Linear(1, 1)

# Loss and optimizer
criterion = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)

# Training loop
for epoch in range(100):
    pred = model(x)
    loss = criterion(pred, y)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

print("Learned weight and bias:", model.weight.item(), model.bias.item())
```

```python
import torch
import matplotlib.pyplot as plt

# Create a random image tensor (1 channel, 28x28)
image = torch.randn(1, 28, 28)

# Remove batch/channel dimension for visualization
plt.imshow(image.squeeze(), cmap='gray')
plt.title("Random Tensor Image")
plt.show()
```

```python
import torch
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Create a random image tensor (1 channel, 28x28)
image = torch.randn(1, 28, 28)

# Prepare the plot
fig, ax = plt.subplots()
ax.imshow(image.squeeze(), cmap='gray')
ax.set_title("Random Tensor Image with Circle")

# Add a circle (center_x, center_y), radius
circle = patches.Circle((14, 14), radius=5, edgecolor='red', facecolor='none', linewidth=2)
ax.add_patch(circle)

plt.show()
```

```python
import torch
import matplotlib.pyplot as plt

# Create a random image tensor (1 channel, 28x28)
image = torch.randn(1, 28, 28)

# Flatten the image
flattened = image.reshape(1, -1)
flattened = flattened.squeeze()  # Remove batch dimension

# Plot the pixel values
plt.plot(flattened.numpy())
plt.title("Flattened Tensor Pixel Intensities")
plt.xlabel("Pixel Index")
plt.ylabel("Intensity")
plt.grid(True)
plt.show()
```

```python
# Reshape back to original image dimensions
reconstructed = flattened.reshape(28, 28)

# Display the image
plt.imshow(reconstructed, cmap='gray')
plt.title("Reconstructed Image from Flattened Tensor")
plt.axis('off')
plt.show()
```
